About the dataset, I choose Q1 as the broad question: What player characteristics and behaviors are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types?
To explore this issue, I plan to use variables related to players' participation as explanatory variables: played_hours and experience from players.csv. And we can get other explanatory variables, which is  total_session and avg_session_duration from the calculation of sessions.csv.The response variable is "subscribe" which the players whether this player has subscribed to the game-related newsletter.
So the specific question is, can player engagement features such as total played hours, number of sessions, and average session duration predict whether a player subscribes to the newsletter?


```r
mean_table <- players %>%
  select(where(is.numeric)) %>%       
  summarise(across(everything(), ~mean(.x, na.rm = TRUE))) %>%  
  pivot_longer(cols = everything(),
               names_to = "Variable",
               values_to = "Mean_Value")  

mean_table$Mean_Value <- round(mean_table$Mean_Value, 2)

kable(mean_table,
      caption = "Mean values of quantitative variables in players.csv",
      align = c("l", "r"))
```
